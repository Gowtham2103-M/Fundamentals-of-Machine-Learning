# Univariate

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as pl
import seaborn as sns
```

```python
df=pd.read_csv("iris.csv")
df
```

```python
df_setosa     = df[df['Species'] == 'Iris-setosa']
df_versicolor = df[df['Species'] == 'Iris-versicolor']
df_virginica  = df[df['Species'] == 'Iris-virginica']
```

```python
plt.figure(figsize=(6, 4))
```

```python


plt.scatter(df_setosa['SepalLengthCm'], np.zeros_like(df_setosa['SepalLengthCm']),
            label='Setosa', color='blue')
plt.scatter(df_versicolor['SepalLengthCm'], np.ones_like(df_versicolor['SepalLengthCm']),
            label='Versicolor', color='yellow')
plt.scatter(df_virginica['SepalLengthCm'], np.full_like(df_virginica['SepalLengthCm'], 2),
            label='Virginica', color='green')

plt.xlabel('Sepal Length (cm)')
plt.yticks([0, 1, 2], ['Setosa', 'Versicolor', 'Virginica'])
plt.title('Univariate Scatter Plot - Sepal Length by Species')
plt.legend()
plt.grid(True, axis='y')
plt.show()
```

```python
# Bivariate scatter
plt.figure(figsize=(6, 4))

plt.scatter(df_setosa['SepalWidthCm'], df_setosa['SepalLengthCm'], 
            color='red', label='Setosa')
plt.scatter(df_versicolor['SepalWidthCm'], df_versicolor['SepalLengthCm'], 
            color='green', label='Versicolor')
plt.scatter(df_virginica['SepalWidthCm'], df_virginica['SepalLengthCm'], 
            color='blue', label='Virginica')

plt.xlabel('Sepal Width (cm)')
plt.ylabel('Sepal Length (cm)')
plt.title('Bivariate Scatter Plot - Sepal Width vs Sepal Length')

plt.legend()
plt.show()
```

```python
sns.FacetGrid(df, hue="Species", height=5) \
   .map(plt.scatter, "PetalLengthCm", "SepalWidthCm") \
   .add_legend()

plt.show()
```

```python
sns.pairplot(df, hue='Species', diag_kind='kde', palette='Set1')
plt.suptitle('Pairplot: Multivariate Relationships (Iris features)', y=1.02)
plt.show()
```
