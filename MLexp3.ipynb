#Least Square Method

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
from sklearn.linear_model import LinearRegression
```

```python
data = pd.read_csv("headbrain.csv")
data
```

```python
data.isnull().sum()
```

```python
def simple_linear_regression_least_squares(data):
    
    X = data["Head Size"].values
    Y = data["Brain Weight"].values

    X_mean = np.mean(X)
    Y_mean = np.mean(Y)
    
    numerator = np.sum((X - X_mean) * (Y - Y_mean))
    denominator = np.sum((X - X_mean) ** 2)
    m = numerator / denominator
    c = Y_mean - m * X_mean

    Y_pred = m * X + c

    ss_total = np.sum((Y - Y_mean) ** 2)
    ss_residual = np.sum((Y - Y_pred) ** 2)
    r2 = 1 - (ss_residual / ss_total)

    print("Simple Linear Regression (Least Squares Method)")
    print("slope(m)",m)
    print("Intercept (c)",c)
    print("R² Score:\n",r2)

    plt.scatter(X, Y, color="green", label="Actual Data")
    plt.plot(X, Y_pred, color="red", label="Regression Line")
    plt.xlabel("Head Size (cm³)")
    plt.ylabel("Brain Weight (grams)")
    plt.title("Simple Linear Regression (Least Squares Method)")
    plt.legend()
    plt.show()

    return m, c, r2, Y_pred

m, c, r2, preds = simple_linear_regression_least_squares(data)
```
